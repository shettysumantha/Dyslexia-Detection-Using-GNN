```python
from dgl.nn import SAGEConv

# Assuming your dataset has 6 features
num_input_features = 6
num_hidden_units = 16
num_output_classes = 2

# Build a two-layer GraphSAGE model
class GraphSAGE(tf.keras.layers.Layer):
    def __init__(self, in_feats, h_feats, num_classes):
        super(GraphSAGE, self).__init__()
        self.conv1 = SAGEConv(in_feats, h_feats, 'mean')
        self.conv2 = SAGEConv(h_feats, num_classes, 'mean')
    
    def call(self, g, in_feat):
        h = self.conv1(g, in_feat)
        h = tf.nn.relu(h)
        h = self.conv2(g, h)
        return h

# Create the GraphSAGE model using the specified dimensions
net = GraphSAGE(num_input_features, num_hidden_units, num_output_classes)
```

```python
import tensorflow as tf

# Convert labels to integer type
labels = tf.cast(labels, dtype=tf.int32)

# Clip label values to be within [0, 1]
labels = tf.clip_by_value(labels, 0, 1)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
loss_fcn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

# ----------- 4. training -------------------------------- #
all_logits = []

# Assuming 'inputs' contains your node features, and 'labels' contains your node labels
for e in range(100):
    
    with tf.GradientTape() as tape:
        tape.watch(inputs)  # Optimize embedding layer also
        
        # Forward pass through the GraphSAGE model
        logits = net(g, inputs)

        # Assuming 'labeled_nodes' contains indices of nodes with labels
        loss = loss_fcn(tf.gather(labels, labeled_nodes),
                        tf.gather(logits, labeled_nodes))

        # Backward pass and optimization
        grads = tape.gradient(loss, net.trainable_weights)
        optimizer.apply_gradients(zip(grads, net.trainable_weights))
        all_logits.append(logits.numpy())
    
    if e % 5 == 0:
        print('In epoch {}, loss: {}'.format(e, loss))
```

```python
pred = tf.argmax(logits, axis=1).numpy()
print('Accuracy', (pred == labels.numpy()).sum().item() / len(pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(labels,pred))
```
